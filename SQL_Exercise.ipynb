## SQL Exercises

### 1. What is the average BACDistance (in Km)? Rename the output column to "avg_bac_distance."

```python
# Write your SQL query in this string
sql = """
SELECT AVG([BACDistance (in Km)])
FROM school_details
"""
# Read the sql query into pandas to view the results
temp_df = pd.read_sql_query(sql, engine)

# Rename the column
temp_df.columns = ['avg_bac_distance']

temp_df.head()
```

```python
# Check that the result is correct

# Write your SQL query in this string
sql = """
SELECT [BACDistance (in Km)]
FROM school_details
"""
# Read the sql query into pandas to view the results
temp_df = pd.read_sql_query(sql, engine)

temp_df.mean()
```

### 2. What is the average BACDistance (in Km) by School Level? Sort the results from highest average to lowest average.

```python
# Write your SQL query in this string
sql = """
SELECT SchoolLevel, AVG([BACDistance (in Km)]) 
FROM school_details
GROUP BY SchoolLevel
ORDER BY AVG([BACDistance (in Km)]) DESC
"""
# Read the sql query into pandas to view the results
temp_df = pd.read_sql_query(sql, engine).dropna()

temp_df.columns = ['SchoolLevel', 'avg_bac_distance']

temp_df.head(10)
```

### 3. Repeat the previous query, but only display results where the average distance is at least 10 km.

```python
# Write your SQL query in this string
sql = """
SELECT SchoolLevel, AVG([BACDistance (in Km)]) 
FROM school_details

GROUP BY SchoolLevel
Having AVG([BACDistance (in Km)]) > 10
ORDER BY AVG([BACDistance (in Km)]) DESC


"""
# Read the sql query into pandas to view the results
temp_df = pd.read_sql_query(sql, engine).dropna()

#rename columns
temp_df.columns = ['SchoolLevel', 'avg_bac_distance']

temp_df.head(10)
```

### 4. Join the two tables together (school_details and school_facility_details). Start by displaying all of the columns, but limit the resulting rows to just 2.

```python
# Write your SQL query in this string
sql = """
SELECT *
FROM school_details d INNER JOIN school_facility_details f
  ON f.EstablishmentCode = d.EstablishmentCode
LIMIT 2
"""
# Read the sql query into pandas to view the results
temp_df = pd.read_sql_query(sql, engine)
temp_df.head()
```

```python
# Display all th ecolumn names
temp_df.columns
```

Notice that the resulting table has two 'index' columns which contains the index of the row from their original tables.

### 5. Now, select just the schools where the SchoolLevel is "PS" that do not have libraries.

```python
# Write your SQL query in this string
sql = """
SELECT *
FROM school_details d INNER JOIN school_facility_details f
  ON f.EstablishmentCode = d.EstablishmentCode
WHERE d.SchoolLevel = 'PS' AND IsLibraryAvailable = 'Not Available'
"""
# Read the sql query into pandas to view the results
temp_df = pd.read_sql_query(sql, engine)
temp_df.head()
```

### 6. Build off the previous query, but order the results alphabetically by name.

```python
# Write your SQL query in this string
sql = """
SELECT *
FROM school_details d INNER JOIN school_facility_details f
  ON f.EstablishmentCode = d.EstablishmentCode
WHERE d.SchoolLevel = 'PS' AND IsLibraryAvailable = 'Not Available'
ORDER BY  EstablishmentName ASC
"""
# Read the sql query into pandas to view the results
temp_df = pd.read_sql_query(sql, engine)
temp_df.head()
```
